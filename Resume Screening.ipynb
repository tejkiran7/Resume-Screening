<b>Cleaning Data:</b><br>
1 URLs,<br>
2 hashtags,<br>
3 mentions,<br>
4 special letters,<br>
5 punctuations:

```python
eng_stopwords = set(stopwords.words('english'))
def cleaning(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = BeautifulSoup(cleanText, 'lxml').get_text()
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    cleanText = ' '.join(e.lower() for e in cleanText.split() if e.lower() not in eng_stopwords)
    return cleanText
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleaning(x))
```

```python
df['Resume'][0]
```

## Dependant variable into Categorical values

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

```python
df.Category.unique()
```

## Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requiredText  = tfidf.transform(df['Resume'])
```

```python
requiredText
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(requiredText, df['Category'], test_size=0.2, random_state=42)
```

```python
X_train.shape,X_test.shape
```

## Now let’s train the model and print the classification report:

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
ypred = clf.predict(X_test)
print(accuracy_score(y_test,ypred))
```
